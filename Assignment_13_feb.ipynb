3) Customization: Inheriting from Exception also allows us to add additional functionality to our custom        exception class if needed, such as custom error messages or additional attributes that provide more          information about the error.

ImportError is a built-in exception class in Python that is raised when a module or package cannot be imported. This can happen for several reasons, including:

1) The module or package does not exist
2) The module or package exists but cannot be found in the search path
3) The module or package exists but contains errors that prevent it from being imported

ModuleNotFoundError is a subclass of ImportError that was introduced in Python 3.6. It is raised when a module or package cannot be found during an attempt to import it. The difference between ModuleNotFoundError and ImportError is that ModuleNotFoundError is more specific and only applies to cases where a module or package cannot be found
