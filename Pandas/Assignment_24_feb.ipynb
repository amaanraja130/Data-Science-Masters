### Ans1)

These are 5 function of pandas library with execution

1) read_csv(): This function is used to read CSV files and create a pandas DataFrame from the data. 

     Here's an example:

```python
import pandas as pd
```

```python
df = pd.read_csv('services.csv')
```

2) head(): This function is used to display the first n rows of a DataFrame. By default, n=5.

    Here's an example:

```python
df.head()
```

3) describe(): This function is used to get statistical information about the data in a DataFrame, such as the mean, standard deviation, and quartiles. 
   
   Here's an example:

```python
df.describe()
```

4) groupby(): This function is used to group the data in a DataFrame based on one or more columns, and then perform an operation on each group.
   
   Here's an example:

```python
d1=df.groupby('audience')
```

```python
d1.mean()
```

5) to_csv(): This function is used to write a pandas DataFrame to a CSV file. 

    Here's an example:

```python
df.to_csv('saved.csv',index=False)
```
