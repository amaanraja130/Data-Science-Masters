### Ans1)

Matplotlib is a data visualization library in Python that allows users to create a wide range of static, animated, and interactive visualizations in Python.
Matplotlib is used for creating visualizations of data and results in various scientific and engineering fields, as well as in business and finance. 
  
  Here are five plots that can be created using the Pyplot module of Matplotlib:

1) Line plot - used to visualize the trend in data over time or other continuous variable. It can be created using the plt.plot() function.

2) Scatter plot - used to visualize the relationship between two variables. It can be created using the plt.scatter() function.

3) Bar plot - used to compare different categories or groups of data. It can be created using the plt.bar() function.

4) Histogram - used to visualize the distribution of a single variable. It can be created using the plt.hist() function.

5) Pie chart - used to visualize the relative proportions of different categories or groups. It can be created using the plt.pie() function.

### Ans2)

A scatter plot is a type of data visualization that displays the relationship between two numerical variables. It is a graph that uses dots or markers to represent the values of two different variables, with one variable plotted along the x-axis and the other variable plotted along the y-axis.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("this is xlabel")
plt.ylabel("this is ylabel")
plt.title("this is title")
```

### Ans3)

The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to divide a figure into a grid of subplots, where each subplot represents a different visualization.

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([50, 20, 40, 20, 60, 70])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([10, 20, 30, 40, 50, 60])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([200, 350, 250, 550, 450, 150])
```

```python
ax=plt.subplot(1,1,1)
ax.plot(x,y)
ax.plot(x1,y1)
ax.plot(x2,y2)
ax.plot(x3,y3)
plt.show()
```

### Ans4)

A bar plot, also known as a bar chart, is a graphical representation of data that displays the frequency, count, or percentage of discrete categories as rectangular bars.

They are useful for displaying categorical data that can be easily understood and interpreted by a wide range of audiences. Bar plots can also be used to compare data across different groups, to track changes over time, or to display the distribution of a single variable.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(6,4))
plt.title("Company Profit")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.barh(company,profit,color="#C12F6B")
plt.show()
```

### Ans5)

A box plot, also known as a box and whisker plot, is a graphical representation of data that displays the distribution of a dataset using five summary statistics - minimum, maximum, median, and quartiles.

Box plots are used to summarize and compare the distribution of a dataset or multiple datasets. They are particularly useful for detecting outliers, comparing distributions across different groups, and identifying asymmetry and skewness in the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
box3 = [box1,box2]
```

```python
plt.boxplot(box3)
plt.show()
```
