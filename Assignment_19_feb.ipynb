1) Lightweight: Flask is a lightweight framework, meaning it doesn't require much boilerplate code or configuration to get started. This makes it easy to learn and use, even for beginners.

2) Flexibility: Flask allows developers to build web applications in their own way, with very few restrictions. This makes it easy to adapt to different types of projects and to add custom functionality as needed.

3) Modular: Flask is designed to be modular, meaning developers can choose which components to use and how to use them. This makes it easy to build small, focused applications that do one thing well.

4) Extensible: Flask is highly extensible, meaning developers can easily add third-party plugins and extensions to add functionality to their applications. This can save a lot of time and effort when building complex web applications.

5) Large Community: Flask has a large and active community of developers who contribute to the framework and provide support for others. This means that developers can easily find help and resources when they need them.

### Ans3)

In Flask, app routing refers to the process of defining URL patterns (also known as endpoints or routes) for handling incoming requests to different parts of your web application.

When a request is made to your Flask application, the application needs to determine which view function should be used to handle that request based on the URL of the request. This is where app routing comes into play - you define URL patterns using Flask's @app.route() decorator, and map them to view functions that generate a response for that URL pattern.

### Ans5)

In Flask, the url_for() function is used for URL building. This function generates a URL to a specific function based on the endpoint name provided.

Here's an example Python code to demonstrate the working of the url_for() function in Flask:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def Class():
    return 'Welcome to the Class!'

@app.route('/profile/<username>')
def profile(username):
    return f'This is {username}'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('Class'))
        print(url_for('profile', username='Amaan'))
```
